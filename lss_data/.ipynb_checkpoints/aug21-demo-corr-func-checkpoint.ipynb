```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123456)
```

```python
x = 20 * np.random.rand(10)
y = 20 * np.random.rand(10)
data = np.array([x,y]).T
data
```

```python
plt.scatter(x,y)
```

```python
distances=[]
k=0
```

```python
distances=[]
k=0
for i in data:
    for j in data:
        d = np.sqrt((i[0]-j[0])**2 + (i[1]-j[1])**2)
        distances.append(d)
        plt.plot([i[0],j[0]],[i[1],j[1]])
    k+=1
```

```python
len(distances)
```

```python
unique_dist = list(set(distances))
print(len(unique_dist))
unique_dist
```

```python
bin_edges = np.linspace(0,20,5)
DD = np.zeros(4)
```

```python
for i in unique_dist:
    k = 0
    while True:
        if bin_edges[k] <= i < bin_edges[k+1]:
            DD[k] = DD[k]+1
            break
        else:
            k += 1
DD
```

```python
DD_hist = plt.hist(unique_dist, bins=bin_edges)
```

```python
DD = np.digitize(unique_dist, bin_edges)
DD
```

```python
_ = plt.hist(DD)
```
