```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 6]

import os
```

```python
abs_path = "/scratch/aew492/research-summer2020_output/1D/"

mock_info = np.load(os.path.join(abs_path, "mock_data/2e-4/cat_L750_n2e-4_z057_patchy_lognormal_rlz400_m-1.000-L_b-0.500.npy"), allow_pickle=True).item()
# use superclustered mock for final plots
```

### Mock Plot

```python
mock_data = mock_info["grad_set"]
L = mock_info["boxsize"]
w_hat = mock_info["w_hat"]

z_max = -50
xy_slice = mock_data[np.where(mock_data[:,2] < z_max)]
```

```python
fig, ax = plt.subplots()
plt.plot(xy_slice[:,0], xy_slice[:,1],',')

# w_hat
a = 0.35*L
plt.arrow(-a, 0, 2*a, 0, color="black", lw=2, head_width = .1*a, head_length=.2*a, length_includes_head=True, zorder=100, label="Input Gradient")

ax.set_aspect("equal")
ax.set_xlabel("x ($h^{-1}\,$Mpc)")
ax.set_ylabel("y ($h^{-1}\,$Mpc)")
plt.legend()
```

### Xi in Patches

```python
patch_info = np.load(os.path.join(abs_path, "patch_data/2e-4/8patches/cat_L750_n2e-4_z057_patchy_lognormal_rlz400_m-1.000-L_b-0.500.npy"), allow_pickle=True).item()
n_patches = patch_info["n_patches"]
r_avg = patch_info["r_avg"]
xi_patches = patch_info["xi_patches"]
xi_patch_avg = patch_info["xi_patch_avg"]
xi_full = patch_info["xi_full"]
patch_centers = patch_info["patch_centers"]
```

```python
fig, ax = plt.subplots()
plt.plot(r_avg, xi_full, color="black", marker=".", label="Full Mock")
plt.plot(r_avg, xi_patch_avg, color="black", alpha=0.5, marker=".", label="Avg. of Patches")

cmap = plt.cm.get_cmap("cool")

for i in range(len(patch_centers)):
    # *figure out how to get exact color from color map
    if patch_centers[i,0] < 0:
        color = "blue"
    elif patch_centers[i,0] >= 0:
        color = "pink"
    plt.plot(r_avg, xi_patches[i], color=color)

ax.axhline(0, color='grey', lw=0.5)
ax.set_xlabel(r'Separation $r$ ($h^{-1}\,$Mpc)')
ax.set_ylabel(r'$\xi$(r)')
ax.set_box_aspect(1)
ax.set_ylim((-0.01, 0.12))
```
