```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_rand = np.random.uniform(low=0.000, high=750.000, size=125342)
y_rand = np.random.uniform(low=0.000, high=750.000, size=125342)
z_rand = np.random.uniform(low=0.000, high=750.000, size=125342)
```

```python
x_rand
```

```python
data_rand = np.transpose(np.array([x_rand, y_rand, z_rand]))
```

```python
data_rand
```

```python
# 2D slice
```

```python
z_max = 70
xy_slice = data_rand[z_rand < z_max] # select rows where z < z_max
```

```python
len(xy_slice)
```

```python
plt.plot(xy_slice[:,0],xy_slice[:,1],',')
plt.xlabel("x (Mpc/h)")
plt.ylabel("y (Mpc/h)")
plt.title("Random Catalog")
```
